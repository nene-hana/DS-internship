A new column review_category, was created by grouping review scores into Poor, Average, and Good to understand how users favor each product.

```python
def categorize_review(review_score):
    if review_score> 4:
        return 'Good'
    elif 2<=review_score <= 4:
        return 'Average'
    else:
        return 'Poor'
df['review_category'] = df['review_score'].apply(categorize_review)  
df.head(5)
```

which product cost high.

```python
df.loc[df.price.idxmax(),'name']
```

Category-wise price statistics were calculated to understand pricing distribution across different product categories.

```python
df.groupby('category').agg(
    maximum_price=('price', 'max'),
    minimum_price=('price', 'min'),
    average_price=('price', 'mean')
).sort_values('average_price', ascending=False)
```

ðŸ“š Business / Domain Questions

1. Which month show peak sales activity?

```python
# Select only the monthly sales columns
monthly_sales = df.loc[:, 'sales_month_1':'sales_month_12'].sum()
monthly_sales.idxmax()
```

*This indicates that Month 10 has the highest sales activity. so, more products should be produced or stocked during this month to meet the increased needs. If sufficient products is not maintained, the platform may fail to satisfy customer needs, leading to  loss of sales.*

2. Which products are overpriced compared to their ratings?

```python
over_priced = df.sort_values(by=['price', 'review_score'], ascending=[False,True]) # Sort by price descending and review_score ascending
over_priced
```

```python
over_priced.head(10)
```

3. Which product have the most customer reviews?

```python
df.loc[df.review_count.idxmax(),"name"]
```

4. Which category has the highest average price?

```python
df.groupby('category')['price'].mean().idxmax()
```

5. Which category shows the highest customer satisfaction?

```python
df.groupby('category')['review_score'].mean().sort_values(ascending=False).head(1)
 
```
