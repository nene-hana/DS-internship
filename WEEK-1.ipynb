3. Use Python basics

```python
top_category = df['category'].value_counts().idxmax()#to find the category with the highest number of products
print(top_category)
```

*Books is the category with the largest number of products listed on the platform. which means  that the company is focused more on providing  books than other categories.*

```python
cheapest_product=df.loc[df.price.idxmin(),"product_name"] #to find the cheapest product in the dataset
print(cheapest_product)
```

*The platform offers products at very low prices, with Product_822 =cheapest.*

```python
sales= df[['sales_month_1','sales_month_2','sales_month_3']].sum() #to calculate total sales for each month
print(sales)
```

```python

sale={"Month 1":sales['sales_month_1'],"Month 2":sales['sales_month_2'],"Month 3":sales['sales_month_3']}#to store the total sales in a dictionary
print(sale)
```

```python
total = 0

for value in df['sales_month_1']: #to calculate total sales in month 1
    total = total + value

print(total)
```

```python
cheap_products=df[df.price<100]#to find all products with price less than 100
print(cheap_products)
```
